```python
with open("/content/dataset.txt", "r") as file:
  # content = file.read()
  # print(content)

  lines = [line for line in file if line.strip()]

  content = "".join(lines)
  print(content)
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts([content])
```

```python
tokenizer.word_index
```

```python
print(len(tokenizer.word_index))
```

```python
input_sequences = []

for sentences in content.split('\n'):
  for sentence in sentences.split('.'):
    # print(sentence)
    # print(tokenizer.texts_to_sequences([sentence])[0])

    tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

    for i in range(1, len(tokenized_sentence)):
      input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen=max_len, padding='pre')
padded_input_sequences
```

```python
X = padded_input_sequences[:, :-1]
X
```

```python
y = padded_input_sequences[:, -1]
y
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y, num_classes=len(tokenizer.word_index)+1)
```

```python
y.shape
```

```python
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()

model.add(Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=100, input_length=max_len-1))
model.add(LSTM(units=128))
model.add(Dense(units=len(tokenizer.word_index)+1, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=10)
```

```python
text = "business"
# tokenize
tokenized_text = tokenizer.texts_to_sequences([text])[0]
# padding
padded_token_text =  pad_sequences([tokenized_text], maxlen=max_len-1, padding='pre')
# predict
pred = model.predict(padded_token_text)
```

```python
pred
```

```python
pred.shape
```

```python
import numpy as np
np.argmax(pred)
```

```python
for word, index in tokenizer.word_index.items():
  if index == np.argmax(pred):
    print(word)
```

```python
text = "occasion"

import time

for i in range(10):
  # tokenize
  tokenized_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text =  pad_sequences([tokenized_text], maxlen=max_len-1, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word, index in tokenizer.word_index.items():
    if index == pos:
      text += " " + word
      print(text)
      time.sleep(2)
```
